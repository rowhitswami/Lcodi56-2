```python
import pandas as pd
df = pd.read_csv('data.csv')
df.columns

# Total number of deaths in all states
state_viz = df[['State', 'Deaths']].groupby('State').sum()
state_viz.sort_values('Deaths', ascending=False)
```

```python
#

without_US = df[df.State.str.contains("United States") == False]
year_viz = without_US[['Year', 'Deaths']].groupby('Year').sum()
year_viz


```
